```python
filename = '/home/sokhorn/sokhorn/dataSet/groceries/products/products.csv'
product_df = pd.read_csv(filename, encoding="utf8", sep=',', engine='python')
product_df
```

```python
df = pd.merge(product_order_df,product_df, how='left', left_on='product_id',right_on='product_id')
df
```

```python
df = df.loc[:, ~df.columns.isin(['add_to_cart_order', 'aisle_id', 'department_id','reordered'])]
df = df.sample(frac=0.3)
```

```python
item_sets = (
    df.groupby(['order_id', 'product_id', ])
    .count().unstack().reset_index().fillna(0)
    .set_index("order_id")
)
item_sets
```
